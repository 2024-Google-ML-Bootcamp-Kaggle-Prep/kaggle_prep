## Encoding categorical features

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
cat=['Gender', 'Vehicle_Age', 'Vehicle_Damage']
for column in cat:
    data[column] = label_encoder.fit_transform(data[column])
```
